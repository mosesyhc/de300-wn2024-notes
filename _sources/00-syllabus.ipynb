## Course Outline

Introduction to the foundation and modern tools of data engineering. Topics include:

-   Introduction to data engineering concepts
-   Exploratory data analysis
-   Data modeling and database management
-   Distributed (cloud) computing
-   Automation of data pipelines
-   A/B testing, *if time permits*
-   Transfer learning, *if time permits*

Northwestern is committed to fostering an academic community respectful and welcoming of persons from all backgrounds. To that end, the [policy on academic accommodations](https://www.northwestern.edu/provost/policies-procedures/classwork-curricular-policies/accommodations-for-religious-holidays.html) for religious holidays stipulates that students will not be penalized for class absences to observe religious holidays. ### Office Hour

Weekly office hours are a dedicated time that the teaching team is available to answer your questions, discuss course content, and generally be of support. If none of the offered times in the syllabus works for you, please email the instructor to schedule an appointment.

### Homework and Lab

You are encouraged to work together on labs and homeworks, but each student must turn in their own assignment and not an exact duplicate of someone else's solutions. The lowest lab will be dropped in calculating your final grades.  

Any late assignment will receive a 20% penalty.  No late assignment is accepted after 72 hours of the due date.

Lab attendance is not mandatory, although you are encouraged to do the labs during the assigned lab time so the teaching team can help you.

### Regrade Policy

In a case where you believe your assignment or exam is misgraded, submit a regrade request via <https://forms.gle/1Kgx1QookPe65MNh6>.

Please observe the following instructions when submitting your request:

-   Please be polite and professional. Mistakes happen, and we will make every effort to ensure everyone gets the grade they earned.
-   Submit **one** form per assignment or exam.
-   All regrade requests should be submitted within **48 hours** from when the assignment or exam is returned to you. After the time period, the regrade request will not be processed.
-   Amount of point deductions is not eligible for regrade.
-   Your entire assignment may be regraded. It is possible to lose points if the regrade uncovers further mistakes.
-   Do not modify an answer and request regrading. You may receive a failing grade for the class, and may be reported for academic dishonesty.

### Ethical Policy

The professor, any TAs, any peer mentors, and the students are expected to abide by all rules and regulations put in place by the Northwestern University, the School of Engineering and the Department of Industrial Engineering and Management Science.

### Generative AI Policy
In this course, you are generally allowed to use Generative Artificial Intelligence (GAI). Any use of GAI should be accompanied by a disclosure at the end of an assignment explaining (1) what you used GAI for; (2) the specific tool(s) you used; and (3) what prompts you used to get the results.
